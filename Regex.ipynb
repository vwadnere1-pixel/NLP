# Mastering Regular Expressions in Python

Regular expressions (often shortened to regex or regexp) are powerful tools for pattern matching and manipulation of strings. They provide a concise and flexible way to identify specific sequences of characters, validate input, and extract information from text.

In this notebook, we'll explore the basics of regular expressions in Python using the built-in `re` module. We'll cover:

1.  **Basic Matching**: How to find simple patterns.
2.  **Special Characters**: Understanding metacharacters like `.`, `*`, `+`, `?`.
3.  **Character Classes**: Using `\d`, `\w`, `\s`, `[]` for specific character sets.
4.  **Quantifiers**: Specifying how many times a character or group should appear.
5.  **Groups and Capturing**: Extracting specific parts of a match.
6.  **Alternation**: Using `|` for 'OR' conditions.

---

## 1. Basic Matching

The `re` module in Python provides functions for working with regular expressions. The most common functions are `re.search()`, `re.match()`, `re.findall()`, and `re.sub()`.

```python
import re

text = "The quick brown fox jumps over the lazy dog."

# re.search(): Scans through string looking for the first location where this regular expression produces a match.
# It returns a match object, or None if no match is found.
match = re.search(r"fox", text)

if match:
    print(f"Found '{match.group()}' at position {match.start()} to {match.end()-1}")
else:
    print("Pattern not found.")
```

```python
text_start = "Python is fun. I love Python."

# re.match(): Checks for a match only at the beginning of the string.
# It also returns a match object or None.
match_start = re.match(r"Python", text_start)
match_fail = re.match(r"fun", text_start)

if match_start:
    print(f"'Python' matches at the start: {match_start.group()}")
else:
    print("'Python' does not match at the start.")

if match_fail:
    print(f"'fun' matches at the start: {match_fail.group()}")
else:
    print("'fun' does not match at the start.")
```

```python
text_multiple = "apple banana cherry apple date"

# re.findall(): Returns all non-overlapping matches of pattern in string, as a list of strings.
all_apples = re.findall(r"apple", text_multiple)
print(f"All occurrences of 'apple': {all_apples}")
```

```python
text_replace = "I like apples and oranges. Apples are healthy."

# re.sub(): Replaces occurrences of a pattern in a string with a replacement string.
new_text = re.sub(r"Apples", "Bananas", text_replace)
print(f"Original: {text_replace}")
print(f"Replaced: {new_text}")

# By default, it replaces all occurrences
new_text_all = re.sub(r"apple", "berry", text_replace.lower())
print(f"Replaced all 'apple' (case-insensitive): {new_text_all}")
```

## 2. Special Characters (Metacharacters)

Regular expressions use special characters (metacharacters) that don't match themselves, but instead signal that some special meaning is intended.

### The Dot (`.`)

The dot `.` matches any single character (except for a newline character by default).

```python
text = "cat, hat, mat, fat, bat, sat, car"

# Match any 3-letter word ending with 'at'
matches = re.findall(r".at", text)
print(f"Words ending with 'at': {matches}")

# Match any character between 'c' and 't'
matches = re.findall(r"c.t", text)
print(f"Words matching 'c.t': {matches}")
```

### The Asterisk (`*`)

The asterisk `*` matches zero or more occurrences of the preceding character or group.

```python
text = "col cooll cooool coooool "

# Match 'co' followed by zero or more 'o's, then 'l'
matches = re.findall(r"co*l", text)
print(f"Matches for 'co*l': {matches}")
```

### The Plus Sign (`+`)

The plus sign `+` matches one or more occurrences of the preceding character or group.

```python
text = "col cooll cooool coooool "

# Match 'co' followed by one or more 'o's, then 'l'
matches = re.findall(r"co+l", text)
print(f"Matches for 'co+l': {matches}")
```

### The Question Mark (`?`)

The question mark `?` matches zero or one occurrence of the preceding character or group (makes it optional).

```python
text = "color colour"

# Match 'color' or 'colour'
matches = re.findall(r"colou?r", text)
print(f"Matches for 'colou?r': {matches}")
```

### Anchors: Caret (`^`) and Dollar Sign (`$`)

The caret `^` matches the start of the string, and the dollar sign `$` matches the end of the string.

```python
text = "Hello world.\nAnother line.\nEnd."

# Match 'Hello' only if it's at the beginning of the string
match_start = re.search(r"^Hello", text)
print(f"'Hello' at start: {match_start.group() if match_start else 'None'}")

# Match 'End.' only if it's at the end of the string
match_end = re.search(r"End.$", text)
print(f"'End.' at end: {match_end.group() if match_end else 'None'}")

# With re.M (MULTILINE) flag, ^ and $ match start/end of each line
match_line_start = re.findall(r"^Another", text, re.M)
print(f"'Another' at line start (MULTILINE): {match_line_start}")
```

## 3. Character Classes

Character classes allow you to match a set of characters. They are defined using square brackets `[]` or predefined shorthand notations.

### Predefined Character Classes

*   `\d`: Matches any digit (0-9).
*   `\D`: Matches any non-digit character.
*   `\w`: Matches any word character (alphanumeric characters plus underscore `_`).
*   `\W`: Matches any non-word character.
*   `\s`: Matches any whitespace character (space, tab, newline, etc.).
*   `\S`: Matches any non-whitespace character.
*   `\b`: Matches a word boundary.
*   `\B`: Matches a non-word boundary.

```python
text = "Phone: 123-456-7890. Email: test@example.com. Price: $19.99."

# Match all digits
digits = re.findall(r"\d", text)
print(f"All digits: {digits}")

# Match sequences of digits (phone number parts)
digit_sequences = re.findall(r"\d+", text)
print(f"Digit sequences: {digit_sequences}")

# Match all word characters
word_chars = re.findall(r"\w", text)
print(f"All word characters: {''.join(word_chars[:20])}...") # show first 20 for brevity

# Match words
words = re.findall(r"\b\w+\b", text)
print(f"All words: {words}")

# Match non-whitespace characters
non_whitespace = re.findall(r"\S", text)
print(f"First 20 non-whitespace chars: {''.join(non_whitespace[:20])}...")
```

### Custom Character Sets (`[]`)

Square brackets `[]` allow you to define your own set of characters to match.

*   `[abc]`: Matches 'a', 'b', or 'c'.
*   `[a-z]`: Matches any lowercase letter.
*   `[A-Z]`: Matches any uppercase letter.
*   `[0-9]`: Matches any digit (same as `\d`).
*   `[a-zA-Z0-9]`: Matches any alphanumeric character.
*   `[^abc]`: Matches any character *except* 'a', 'b', or 'c' (negation).

```python
text = "apple, banana, cherry, date, grape"

# Match any vowel
vowels = re.findall(r"[aeiouAEIOU]", text)
print(f"All vowels: {vowels}")

# Match any character that is NOT a vowel
non_vowels = re.findall(r"[^aeiouAEIOU\s,]", text) # Exclude spaces and commas too
print(f"All non-vowels (excluding spaces/commas): {non_vowels}")

# Match words starting with 'b' or 'c'
start_bc = re.findall(r"\b[bc]\w+\b", text)
print(f"Words starting with 'b' or 'c': {start_bc}")
```

## 4. Quantifiers

Quantifiers specify how many occurrences of the preceding character or group must be present for a match.

*   `{n}`: Exactly `n` occurrences.
*   `{n,}`: `n` or more occurrences.
*   `{n,m}`: Between `n` and `m` occurrences (inclusive).

```python
text = "123 1234 12345 12 1"

# Match exactly 3 digits
three_digits = re.findall(r"\d{3}", text)
print(f"Exactly 3 digits: {three_digits}")

# Match 3 or more digits
three_or_more_digits = re.findall(r"\d{3,}", text)
print(f"3 or more digits: {three_or_more_digits}")

# Match between 2 and 4 digits
two_to_four_digits = re.findall(r"\d{2,4}", text)
print(f"Between 2 and 4 digits: {two_to_four_digits}")
```

## 5. Groups and Capturing

Parentheses `()` are used to create capturing groups. This allows you to treat multiple characters as a single unit and to extract specific parts of a matched string.

```python
text = "Name: John Doe, Age: 30. Name: Jane Smith, Age: 25."

# Find all full names
full_names = re.findall(r"Name: (\w+ \w+)", text)
print(f"Full names: {full_names}")

# Find all ages
ages = re.findall(r"Age: (\d+)", text)
print(f"Ages: {ages}")

# Find both name and age using multiple groups
name_age_pairs = re.findall(r"Name: (\w+ \w+), Age: (\d+)", text)
print(f"Name and age pairs: {name_age_pairs}")

# Accessing groups from a match object
match = re.search(r"Name: (\w+ \w+), Age: (\d+)", text)
if match:
    print(f"First name found: {match.group(1)}") # Group 1 is the first (\w+ \w+)
    print(f"First age found: {match.group(2)}") # Group 2 is the second (\d+)
    print(f"Entire match: {match.group(0)}") # Group 0 is the entire match
```

## 6. Alternation (`|`)

The pipe symbol `|` acts like an 'OR' operator, allowing you to match one of several patterns.

```python
text = "cat dog bird fish fox"

# Match 'cat' or 'dog'
matches = re.findall(r"cat|dog", text)
print(f"Matches for 'cat|dog': {matches}")

# Match 'bird' or 'fish' or 'fox'
matches = re.findall(r"bird|fish|fox", text)
print(f"Matches for 'bird|fish|fox': {matches}")

# Using groups with alternation
matches = re.findall(r"f(ox|ish)", text) # Matches 'fox' or 'fish'
print(f"Matches for 'f(ox|ish)': {matches}")
```
