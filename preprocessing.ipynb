```python
import re
def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    text = pattern.sub('', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

```python
def chat_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in slang_dict:
            new_text.append(slang_dict[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
import nltk
from nltk.corpus import stopwords
```

```python

nltk.download('stopwords')
stopwords.words('english')
```

```python
stop_words = set(stopwords.words('english'))

def remove_stopwords1(text):
    new_text = []

    for word in text.split():
        if word.lower() not in stop_words:
            new_text.append(word)

    return " ".join(new_text)
```

```python
def remove_stopwords(text):
    return " ".join([word for word in text.split() if word.lower() not in stop_words])
```

```python
text = "This is a very simple example to demonstrate how we remove stopwords from a sentence"
```

```python
remove_stopwords1(text)
```

```python
remove_stopwords(text)
```

```python
text = "I love this movie рҹҳҚрҹ”Ҙ but the ending was sad рҹҳўрҹ’”"
```

```python
import re

def remove_emoji(text):
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map
        u"\U0001F1E0-\U0001F1FF"  # flags
        "]+", flags=re.UNICODE)

    return emoji_pattern.sub(r'', text)
```

```python
print(remove_emoji(text))
```

```python
# Step 1: Imports
import nltk
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag, word_tokenize

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')


lemmatizer = WordNetLemmatizer()

#convert NLTK POS вҶ’ WordNet POS
def get_wordnet_pos(tag):
    if tag.startswith('J'):   # Adjective
        return wordnet.ADJ
    elif tag.startswith('V'): # Verb
        return wordnet.VERB
    elif tag.startswith('N'): # Noun
        return wordnet.NOUN
    elif tag.startswith('R'): # Adverb
        return wordnet.ADV
    else:
        return wordnet.NOUN   # Default

def advanced_lemmatization(text):
    # Tokenize
    words = word_tokenize(text)

    # POS tagging
    tagged_words = pos_tag(words)

    # Lemmatize with POS
    lemmatized_words = [
        lemmatizer.lemmatize(word, get_wordnet_pos(tag))
        for word, tag in tagged_words
    ]

    return " ".join(lemmatized_words)
```

```python
advanced_lemmatization(text)
```
