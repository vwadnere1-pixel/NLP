```python
import nltk
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

nltk.download('punkt')
```

```python
text = """
Elon Musk founded Tesla to accelerate the world's transition to sustainable energy.
Electric vehicles were once considered inefficient and expensive.
Tesla changed this perception by building high-performance electric cars.
Today, Tesla is a leader in electric vehicle innovation and clean energy solutions.
The company continues to push boundaries in battery technology and automation.
"""
```

```python
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(text)
print(sentences)
```

```python
vectorizer = TfidfVectorizer(stop_words='english')

X = vectorizer.fit_transform(sentences)
```

```python
sentence_scores = np.sum(X.toarray(), axis=1)

for i, score in enumerate(sentence_scores):
    print(f"Sentence {i}: Score = {score}")
```

```python
N = 2

top_indices = np.argsort(sentence_scores)[-N:]
top_indices = sorted(top_indices)
summary = " ".join([sentences[i] for i in top_indices])
print(summary)
```

```python
def summarize_text(text, num_sentences=2):
    sentences = sent_tokenize(text)

    vectorizer = TfidfVectorizer(stop_words='english')
    X = vectorizer.fit_transform(sentences)
    scores = np.sum(X.toarray(), axis=1)
    top_indices = np.argsort(scores)[-num_sentences:]
    top_indices = sorted(top_indices)

    summary = " ".join([sentences[i] for i in top_indices])

    return summary
```

```python
print(summarize_text(text, 2))
```

```python
import numpy as np
import nltk
import networkx as nx

from nltk.tokenize import sent_tokenize, word_tokenize
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

from gensim.models import Word2Vec

nltk.download('punkt')
```

```python
text = """
Elon Musk founded Tesla to accelerate the world's transition to sustainable energy.
Electric vehicles were once considered inefficient and expensive.
Tesla changed this perception by building high-performance electric cars.
Today, Tesla is a leader in electric vehicle innovation and clean energy solutions.
The company continues to push boundaries in battery technology and automation.
Electric mobility is becoming increasingly important due to climate change concerns.
"""
```

```python
sentences = sent_tokenize(text)

def preprocess(sentence):
    return [word.lower() for word in word_tokenize(sentence) if word.isalnum()]

processed_sentences = [preprocess(s) for s in sentences]
```

```python
w2v_model = Word2Vec(sentences=processed_sentences, vector_size=100, window=5, min_count=1, workers=4)
```

```python
def sentence_vector(sentence):
    words = preprocess(sentence)
    vec = np.zeros(100)

    count = 0
    for word in words:
        if word in w2v_model.wv:
            vec += w2v_model.wv[word]
            count += 1

    if count != 0:
        vec = vec / count

    return vec

sentence_vectors = np.array([sentence_vector(s) for s in sentences])
```

```python
similarity_matrix = cosine_similarity(sentence_vectors)
```

```python
graph = nx.from_numpy_array(similarity_matrix)

scores = nx.pagerank(graph)
```

```python
def remove_redundancy(selected, threshold=0.7):
    final_sentences = []

    for idx in selected:
        add = True
        for f_idx in final_sentences:
            sim = cosine_similarity(
                [sentence_vectors[idx]],
                [sentence_vectors[f_idx]]
            )[0][0]

            if sim > threshold:
                add = False
                break

        if add:
            final_sentences.append(idx)

    return final_sentences
```

```python
ranked_sentences = sorted(((scores[i], i) for i in range(len(sentences))), reverse=True)

top_n = 3
selected_indices = [idx for (_, idx) in ranked_sentences[:top_n]]
selected_indices = remove_redundancy(selected_indices)
selected_indices = sorted(selected_indices)
summary = " ".join([sentences[i] for i in selected_indices])
print(summary)
```

```python
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(sentences)

lda = LatentDirichletAllocation(n_components=2, random_state=42)
lda.fit(X)
topic_scores = lda.transform(X)
topic_strength = np.max(topic_scores, axis=1)
top_indices = np.argsort(topic_strength)[-3:]
top_indices = sorted(top_indices)

lda_summary = " ".join([sentences[i] for i in top_indices])

print(lda_summary)
```
